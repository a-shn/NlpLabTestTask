```
os.environ['MALLET_HOME'] = '/content/mallet-2.0.8'
mallet_path = '/content/mallet-2.0.8/bin/mallet' # you should NOT need to change this 
```

```
import pyLDAvis.gensim as gensimvis
import pyLDAvis
import gensim
import csv
import logging

from gensim.corpora import Dictionary
from gensim.models.wrappers import LdaMallet

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

tweets = []
with open('data/clear_covid_tweets.csv', newline='') as csvfile:
	reader = csv.DictReader(csvfile)
	for tweet in reader:
		tweets.append(tweet['text'].split(' '))

dictionary = Dictionary(tweets)
corpus = [dictionary.doc2bow(tweet) for tweet in tweets]
print('before model')
model = gensim.models.wrappers.ldamallet.malletmodel2ldamodel(
	LdaMallet(mallet_path, corpus=corpus, num_topics=50, id2word=dictionary))
print('after model, before vis')

vis_data = gensimvis.prepare(model, corpus, dictionary)
print('after vis, before save')
pyLDAvis.save_html(vis_data, 'lda.html')
print('after save')
```
